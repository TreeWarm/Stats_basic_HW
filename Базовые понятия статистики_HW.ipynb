### Выбросы

```python
def outlier(df_horse, param):
    
    '''
        подсчет выбросов методом межквартильного размаха
        функция возвращает выбросы
        значения NaN в param-столбце удалены
        
    '''
    
    q1 = df_horse[param].quantile(0.25)
    q3 = df_horse[param].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr) 
    upper_bound = q3 + (1.5 * iqr)
    

    outliers = df_horse[~df_horse[param].between(lower_bound, upper_bound, inclusive=True)]


    return outliers[[param]].dropna()  
#     return outliers[param].dropna().tolist() вывод выбросов в виде списка 
#     return outliers.dropna() вывод выбросов в виде кадра данных с пропуском NaN
```

```python
out_rect = outlier(df_horse, 'rectal temperature')
out_rect                   
```

```python
out_pulse = outlier(df_horse, 'pulse')
out_pulse  
```

```python
out_resp = outlier(df_horse, 'respiratory rate')
out_resp
```

```python
# animals with painful lesions or suffering from circulatory shock
# may have an elevated heart rate

fill_horse['pulse'].fillna(fill_horse.groupby(['pain', 'Age'])['pulse'].transform('median'), inplace=True)
```

```python
fill_horse['pain'].fillna(fill_horse.groupby(pd.cut(fill_horse['pulse'], [0,40,185]))['pain'].transform('median'), inplace=True)

# is a reflection of the heart condition: 30 -40 is normal for adults
# - rare to have a lower than normal rate although athletic horses
# may have a rate of 20-25
# предполагается, что боль зависит от пульса , пульс разбиваем на 2 диапазона - меньше 40 и до максимума
```

```python
fill_horse['respiratory rate'].fillna(fill_horse['respiratory rate'].mean(), inplace=True)
# - usefulness is doubtful due to the great fluctuations
# заменю на среднее значение
```

```python
fill_horse['rectal temperature'].fillna(fill_horse.groupby(pd.cut(fill_horse['pulse'],[0,40,185]))['rectal temperature'].transform('median'), inplace=True)

# предполагаю, что температура зависит от пульса
```
